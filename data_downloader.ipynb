### Теперь у нас есть 1 тег -> 1 текст

Для ML это мало, нам нужно 1 тег -> 100 текстов.

Для того, чтобы справиться с этой проблемой мы возьмём каждое предложение в тексте и сопаставим его тегу.

```python
import re

new_tags = []
new_text = []

for tag,text in zip(tags, texts):
    for sentence in re.findall('[A-Z]{1}[^\.]{5,}', text):
        new_tags.append(tag.lower())
        new_sentense = ''
        for word in re.findall('[A-Za-z]* ', sentence):
            new_sentense+=word.lower()
        new_text.append(new_sentense)
    print(len(new_tags), len(new_text))
```

```python
#Сохранимся и пошафлим дату для ML
import numpy as np

wiki = pd.DataFrame()

wiki['tags'] = [i for i in new_tags]
wiki['text'] = [i for i in new_text]

wiki = wiki.reindex(np.random.permutation(wiki.index))

wiki.to_csv('wiki_new_data.csv')
```
